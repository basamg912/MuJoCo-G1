```python
import pandas as pd
```

```python
all_dfs = []
for tag in ea.Tags()["scalars"]:
    events = ea.Scalars(tag)
    df = pd.DataFrame([(e.step, e.value) for e in events], columns=["step", "value"])
    df["tag"] = tag
    all_dfs.append(df)

pd.concat(all_dfs).to_csv("training_log.csv", index=False)
```
